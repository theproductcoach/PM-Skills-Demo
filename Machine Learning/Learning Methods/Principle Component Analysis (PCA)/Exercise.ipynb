```python
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv("heart.csv")

df.head()

df.shape
```

```python
numeric_columns = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
```

```python
outliers = pd.DataFrame()

for col in numeric_columns:
    df[col + '_Zscore'] = zscore(df[col])
    # Find rows where the Z-score is greater than 3 or less than -3
    outliers_col = df[(df[col + '_Zscore'] > 3) | (df[col + '_Zscore'] < -3)]
    # Append these rows to the outliers DataFrame
    outliers = pd.concat([outliers, outliers_col], axis=0).drop_duplicates()

outliers
```

```python
outlier_indices = outliers.index
```

```python
df_cleaned = df.drop(outlier_indices)

df_cleaned.shape
```
